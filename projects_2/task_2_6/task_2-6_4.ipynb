```python
print("Вы находите камень на котором нарисованы три стрелки и одинокий вопрос:")
way = input("Куда пойдешь?: ").strip().lower()

if way == "направо" or way == "вправо":
    print("Вас ждет потрясающий ужин, поздравляю!")
elif way == "налево" or way == "влево":
    print("Вы снова наткнулись на камень, на котором три стрелки, чтобы это могло значить?")
elif way == "вперёд" or way == "прямо":
    print("Эх, богатырь, мне тебя жаль. Тебя уничтожил дракон...")
else:
    print("Нужно сделать выбор!")
```
